```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

```

```python
# 1. Load Dataset [cite: 16, 18]
iris = load_iris()
X = pd.DataFrame(iris.data, columns=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
y = iris.target

# 2. Train-Test Split 
# Training on 50% and testing on 100% as requested to simulate "worse" performance
X_train, _, y_train, _ = train_test_split(X, y, train_size=0.5, random_state=42, stratify=y)
X_test, y_test = X, y 

# 3. Feature Scaling 
# Optimization: KNN and Logistic Regression are distance/gradient-based and require scaling.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

# 4. Initialize Models [cite: 11, 12, 13]
models = {
    "K-Nearest Neighbours": KNeighborsClassifier(n_neighbors=5),
    "Logistic Regression": LogisticRegression(max_iter=200),
    "Naive Bayes": GaussianNB()
}

# 5. Training and Evaluation [cite: 14]
results = {}

print("--- Model Performance Report ---")
for name, model in models.items():
    # Optimization: Use scaled data for KNN and Logistic, raw data is fine for Naive Bayes
    if name == "Naive Bayes":
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)
    else:
        model.fit(X_train_scaled, y_train)
        predictions = model.predict(X_test_scaled)
    
    accuracy = accuracy_score(y_test, predictions)
    results[name] = accuracy
    
    print(f"\nModel: {name}")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Classification Report:\n{classification_report(y_test, predictions, target_names=iris.target_names)}")

# 6. Identify the Best Algorithm [cite: 14]
best_model = max(results, key=results.get)
print(f"\nSummary: The best performing model for this task is {best_model}.")
```
