```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df.isnull().sum()
```

```python
titanic_df.head()
# titanic_df['age'].plot(kind='box') # checking outliers in age
```

```python
# clean_and _handle data
features =titanic_df[["pclass","sex","age","fare","embarked"]]
target = titanic_df["survived"]
imputer = SimpleImputer(strategy = "median") 
titanic_df[["age"]] = imputer.fit_transform(titanic_df[["age"]])
imputer = SimpleImputer(strategy = "most_frequent") 
titanic_df[['embarked']] = imputer.fit_transform(titanic_df[['embarked']])
titanic_df.info()
```
