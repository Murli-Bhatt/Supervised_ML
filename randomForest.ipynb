```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df.isnull().sum()
```

```python
titanic_df.head()
# titanic_df['age'].plot(kind='box') # checking outliers in age
```

```python
# clean_and _handle data
features =["pclass","sex","age","fare","embarked"]
target = ["survived"]
imputer = SimpleImputer(strategy = "mean") 
titanic_df[["age"]] = imputer.fit_transform(titanic_df[["age"]])
imputer = SimpleImputer(strategy = "most_frequent") 
titanic_df[['embarked']] = imputer.fit_transform(titanic_df[['embarked']])
titanic_df.info()
```

```python

# Copy data
x = titanic_df[features].copy()
y = titanic_df["survived"].copy()

# Label Encoding
le = LabelEncoder()
x["sex"] = le.fit_transform(x["sex"])
x["embarked"] = le.fit_transform(x["embarked"])

x.head()
```

```python
# train test split
x_train ,x_test,y_train,y_test =train_test_split(x,y,test_size = 0.2,random_state=42)
# model
clasifier = DecisionTreeClassifier(max_depth = 4)
clasifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clasifier.predict(x_test)
y_train_pred = clasifier.predict(x_train)

print("Training Accuracy : ",accuracy_score(y_train,y_train_pred))
print("testing Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier 
rf = RandomForestClassifier(max_depth = 4,
                            n_estimators = 401,
                            oob_score = True
                           )
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
print("OOB score : ",rf.oob_score_)
print("accuracy score : ",accuracy_score(y_test,y_pred))
```
